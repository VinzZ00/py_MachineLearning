```python
import pandas as pd
```

```python
weather = pd.read_csv("https://milliams.com/courses/data_analysis_python/cetml1659on.txt", skiprows=4, delim_whitespace = True)
weather[weather < 0] = None
weather.tail()
```

```python
print((weather["YEAR"].max() + weather["YEAR"].min())/2)
```

```python
weather["YEAR"] = pd.cut(weather["YEAR"], 2, labels = ["High","Low"])
```

```python
weather
```

```python
from numpy import random
```

```python
div = random.uniform(0,1,len(weather))
Div = []
for x in div :
    if x < 0.75 :
        Div.append(False)
    else :
        Div.append(True)
weather["Devider"] = Div
```

```python
Train, test = weather[weather["Devider"] == False], weather[weather["Devider"] == True]

print(len(Train), len(test))
```

```python
Train = Train.dropna()
datafeature = Train.columns[:13]
feature = Train.columns[:12]
Train = Train[datafeature]
Train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_jobs = 2)
rfc.fit(Train[feature], Train["YEAR"])
```
